f) ¿Cuantos clientes hay de genero femenino?

h) Mostrar solo a los clientes entre 18 y 25 años que tengan un ingreso anual mayor a 15k.


```python
#Creamos los filtros tal como vimos en clases.
mask1 = Clientes['Age'] >= 18
mask2 = Clientes['Age'] <= 25
mask3 = Clientes['Annual Income (k$)'] > 15.0

#Aplicamos los filtros para obtener la seleccion deseada
filtrado = Clientes[mask1 & mask2 & mask3]

print(filtrado[['Age', 'Annual Income (k$)']])
```

m) Agrupar por grupos de edad (ejemplo: <30, 30-50, >50) y obtener la media de ingreso.


```python
#Creamos los filtros, en este caso como lo sugerido en el inciso
GruposdeEdad1 = Clientes['Age'] < 30
GruposdeEdad2 = (Clientes['Age'] >= 30) & (Clientes['Age'] <= 50)
GruposdeEdad3 = Clientes['Age'] > 50

#Calculamos promedio de cada grupo
GruposdeEdadMean = Clientes[GruposdeEdad1]['Annual Income (k$)'].mean()
GruposdeEdadMean2 = Clientes[GruposdeEdad2]['Annual Income (k$)'].mean()
GruposdeEdad3Mean = Clientes[GruposdeEdad3]['Annual Income (k$)'].mean()

print(f'El ingreso medio del grupo 1: {GruposdeEdadMean}')
print(f'El ingreso medio del grupo 2: {GruposdeEdadMean2}')
print(f'El ingreso medio del grupo 3: {GruposdeEdad3Mean}')

```

# Creacion de nuevas columnas

n) Crear una columna llamada Income_Category que tenga:
 - Low si Annual Income < 20
 - Medium si 20 <= Income < 40
 - High si Income >= 40

```python
Income_Category = [] #Creamos una lista vacia para los datos de la nueva columna

#recorremos la columna de ingresos anuales y asi poder clasificarlos en las distintas categorias
for income in Clientes['Annual Income (k$)']:
  if income < 20:
    Income_Category.append('Low')
  elif 20 <= income < 40:
      Income_Category.append('Medium')
  else:
        Income_Category.append('High')

#Creamos la nueva columna
Clientes['Income_Category'] = Income_Category

print(Clientes) #Se puede ver la columna sumada en el final
```

o) Crear una columna Age_Group con categorías: Joven (<30), Adulto (30-60),
Mayor (>60)**texto en negrita**

```python
Age_group = [] #Creamos una lista vacia para los datos de la nueva columna

# asi como en el punto anterior, recorremos la columna de age para poder clasificar
for age in Clientes['Age']:
 if age < 30:
  Age_group.append('Joven')
 elif 30 <= age < 60:
  Age_group.append('Adulto')
 else:
  Age_group.append('Mayor')

#Creamos la nueva columna
Clientes['Age_group'] = Age_group

print(Clientes) #Se puede ver la columna sumada en el final
```
